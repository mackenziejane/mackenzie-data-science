```python
## Histogram on converted subscribers
df['users.is_converted'].plot(kind='hist', bins=2)
```

```python
pd.crosstab(df['leads.marketing_channel'],df['users.is_converted'], margins = True)
```

```python
df['leads.marketing_channel'].value_counts().plot(kind='bar')
```

```python
conv_leads_marketing_channel = df[['leads.marketing_channel','users.is_converted']]
conv_leads_marketing_channel = conv_leads_marketing_channel[(conv_leads_marketing_channel['users.is_converted']==1)]
conv_leads_marketing_channel['leads.marketing_channel'].value_counts().plot(kind='bar')
```

```python
pd.crosstab(df['users.marketing_channel'],df['users.is_converted'], margins = True)
```

```python
df['users.marketing_channel'].value_counts().plot(kind='bar')
```

```python
conv_users_marketing_channel = df[['users.marketing_channel','users.is_converted']]
conv_users_marketing_channel = conv_users_marketing_channel[(conv_users_marketing_channel['users.is_converted']==1)]
conv_users_marketing_channel['users.marketing_channel'].value_counts().plot(kind='bar')
```
